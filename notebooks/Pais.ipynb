# Perguntas de Negócio

### 1. Qual o nome do país que possui mais cidades registradas?

```python
cols = ['country_name', 'city']

paises_agrupados = df.loc[ : , cols].groupby('country_name')['city'].count().sort_values(ascending=False).reset_index(name='Quantidade de Cidades').iloc[0]
nome = paises_agrupados['country_name']
qtd_cidades = paises_agrupados['Quantidade de Cidades']

print(f'O país com mais cidades registradas é a {nome}, com {qtd_cidades} registros')
```

### 2. Qual o nome do país que possui mais restaurantes registrados?

```python
cols = ['country_name', 'restaurant_id']

paises_agrupados = df.loc[ : , cols].groupby('country_name')['restaurant_id'].count().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').iloc[0]
nome = paises_agrupados['country_name']
qtd_paises = paises_agrupados['Quantidade de Restaurantes']

print(f'O país com mais restaurantes registradas é a {nome}, com {qtd_paises} registros')
```

### 3. Qual o nome do país que possui mais restaurantes com o nível de categoria igual a Gourmet registrados?

```python
contagem_paises = df.loc[df['price_category'] == 'gourmet' , ['country_name' , 'restaurant_id']].groupby('country_name')['restaurant_id'].nunique().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').iloc[0]
nome = contagem_paises['country_name']
qtd_restaurantes = contagem_paises['Quantidade de Restaurantes']

print(f'O país com mais restaurantes de nível Gourmet é {nome}, com {qtd_restaurantes} registrados')
```

### 4. Qual o nome do país que possui a maior quantidade de tipos de culinária distintos?

```python
paises_agrupados = df.loc[ : , ['country_name', 'cuisines']].groupby('country_name')['cuisines'].nunique().sort_values(ascending=False).reset_index(name='Quantidade de Tipos Culinários').iloc[0]
nome = paises_agrupados['country_name']
qtd_tipos = paises_agrupados['Quantidade de Tipos Culinários']

print(f'O país com a maior quantidade de tipos culinários é {nome}, com {qtd_tipos} tipos')
```

### 5. Qual o nome do país que possui a maior quantidade de avaliações feitas?

```python
paises_agrupados = df.loc [ : , ['country_name', 'votes']].groupby('country_name')['votes'].sum().sort_values(ascending=False).reset_index(name='Quantidade de Avaliacoes').iloc[0]
nome = paises_agrupados['country_name']
qtd_avaliacoes = paises_agrupados['Quantidade de Avaliacoes']

print(f'O país com a maior quantidade de avaliações é {nome}, com {qtd_avaliacoes} no total')
```

# Funções de Gráficos

```python
# Quantidade de restraurantes registrados por país

def agrupar_restaurante_pais(df):
  df_restaurantes = df.groupby('country_name')['restaurant_id'].nunique().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes')
  fig = px.bar(df_restaurantes, x='country_name', y='Quantidade de Restaurantes', text='Quantidade de Restaurantes', color='country_name', labels={'country_name': 'País'})

  fig.update_layout(xaxis_title='Países', showlegend=False)
  
  return fig

agrupar_restaurante_pais(df)
```

```python
# Quantidade de cidades registradas por país

def agrupar_cidade_pais(df):
  df_cidade_pais = df.groupby('country_name')['city'].nunique().sort_values(ascending=False).reset_index(name='Quantidade de Cidades')
  fig = px.bar(df_cidade_pais, x='country_name', y='Quantidade de Cidades', text='Quantidade de Cidades', color='country_name', labels={'country_name': 'País'})
  
  fig.update_layout(xaxis_title='Países', showlegend=False)

  return fig

agrupar_cidade_pais(df)
```

```python
# Média de avaliações por pais

def avaliacao_media_pais(df):
  df_media_pais = df.groupby('country_name')['votes'].mean().sort_values(ascending=False).reset_index(name='Media de Avaliacoes').round(2)
  fig = px.bar(df_media_pais, x='country_name', y='Media de Avaliacoes', text='Media de Avaliacoes', color='country_name', labels={'country_name': 'País'})
  
  fig.update_layout(xaxis_title='Países', showlegend=False)

  return fig

avaliacao_media_pais(df)
```

```python
# Média de preço de um prato pra duas pessoas por país

def media_prato_dois(df):
  df_media_prato_dois = df.groupby('country_name')['average_cost_for_two'].mean().sort_values(ascending=False).reset_index(name='Media Prato para Dois').round(2)
  fig = px.bar(df_media_prato_dois, x='country_name', y='Media Prato para Dois', text='Media Prato para Dois')

  fig.update_layout(xaxis_title='Países')
  

  return fig

media_prato_dois(df)
```
