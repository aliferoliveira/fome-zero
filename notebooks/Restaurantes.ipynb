# Perguntas de Negócio

### 1. Qual o nome do restaurante que possui a maior quantidade de avaliações?

```python
restaurante_qtd_av = df.groupby('restaurant_name')['votes'].sum().sort_values(ascending=False).reset_index(name='Maior Quantidade de Avaliacoes').iloc[0]
nome = restaurante_qtd_av['restaurant_name']
qtd_avaliacoes = restaurante_qtd_av['Maior Quantidade de Avaliacoes']

print(f'O restaurante com a maior quantidade de avaliacoes é {nome}, com {qtd_avaliacoes} registros')
```

### 2. Qual o nome do restaurante com a maior nota média?

```python
restaurante_maior_media = (df.groupby('restaurant_name')
                           .agg(
                            maior_quantidade_avaliacoes=('aggregate_rating', 'mean'),
                            menor_id=('restaurant_id', 'min')
                        ).sort_values(by=[ 'maior_quantidade_avaliacoes', 'menor_id' ] , ascending=[False, True]).reset_index().iloc[0]
)
nome = restaurante_maior_media['restaurant_name']
media_av = restaurante_maior_media['maior_quantidade_avaliacoes']

print(f'O restaurante com a maior nota média é {nome}, com {media_av} de nota')
```

### 3. Qual o nome do restaurante que possui o maior valor de uma prato para duas pessoas?

```python
restaurante_valor_dois = df.groupby('restaurant_name')['average_cost_for_two'].max().sort_values(ascending=False).reset_index(name='Maior Valor de Prato').iloc[0]
nome = restaurante_valor_dois['restaurant_name']
maior_valor = restaurante_valor_dois['Maior Valor de Prato']

print(f'O restaurante com o prato mais caro é {nome}, sendo na quantia de {maior_valor}')
```

### 4. Qual o nome do restaurante de tipo de culinária brasileira que possui a menor média de avaliação?


```python
df_brasil = (df.loc[(df['country_name'] == 'Brazil') & (df['aggregate_rating'] > 0) , ['restaurant_name', 'aggregate_rating', 'restaurant_id']].groupby('restaurant_name')
             .agg(
                 media_avaliacoes=('aggregate_rating', 'mean'),
                 menor_id=('restaurant_id', 'min')
             ).sort_values(by=['media_avaliacoes', 'menor_id'] , ascending=[True, True]).reset_index().iloc[0])
nome = df_brasil['restaurant_name']
media_av = df_brasil['media_avaliacoes']

print(f'O restaurante com culinária brasileira e com a menor nota média é o {nome}, com média {media_av}')
```

### 5. Qual o nome do restaurante de tipo de culinária brasileira, e que é do Brasil, que possui a maior média de avaliação?

```python
df_brasil = (df.loc[(df['country_name'] == 'Brazil') & (df['aggregate_rating'] > 0) , ['restaurant_name', 'aggregate_rating', 'restaurant_id']].groupby('restaurant_name')
             .agg(
                 media_avaliacoes=('aggregate_rating', 'mean'),
                 menor_id=('restaurant_id', 'min')
             ).sort_values(by=['media_avaliacoes', 'menor_id'] , ascending=[False, True]).reset_index().iloc[0])
nome = df_brasil['restaurant_name']
media_av = df_brasil['media_avaliacoes']

print(f'O restaurante com culinária brasileira e com a maior nota média é o {nome}, com média {media_av}')
```

### 6. Os restaurantes que aceitam pedido online são também, na média, os restaurantes que mais possuem avaliações registradas?

```python
media_online = df.loc[df['has_online_delivery'] == 1, 'votes'].mean().round(1)
media_offline = df.loc[df['has_online_delivery'] == 0, 'votes'].mean().round(1)

print(
    f'Os restaurantes com pedidos online têm média de {media_online} avaliações, '
    f'enquanto os sem pedidos online têm {media_offline}'
)
print(f'\nSim, os restaurantes que aceitam pedidos online são os que mais possuem avaliações registradas')
```

### 7. Os restaurantes que fazem reservas são também, na média, os restaurantes que possuem o maior valor médio de um prato para duas pessoas?


```python
media_reserva = df.loc[df['has_table_booking'] == 1 , 'average_cost_for_two'].mean().round(1)
media_sem_reserva = df.loc[df['has_table_booking'] == 0, 'average_cost_for_two'].mean().round(1)

print(
    f'Os restaurantes com pedidos reserva têm valor médio de um prato de duas pessoas de {media_reserva}, '
    f'enquanto os sem reserva tem média de valor de {media_sem_reserva}'
)
print(f'\nSim, os restaurantes que fazem reserva possuem o maior valor médio de um prato para duas pessoas')
```

# Funções de Gráficos


```python
def grafico_online_offline(df):

    media_online = df.loc[df['has_online_delivery'] == 1, 'votes'].mean().round(1)
    media_offline = df.loc[df['has_online_delivery'] == 0, 'votes'].mean().round(1)

    grafico = pd.DataFrame({
        'Tipo': ['Online', 'Offline'],
        'Média de Avaliaçoes': [media_online, media_offline]
    } )

    fig = px.bar(grafico, x='Tipo', y='Média de Avaliaçoes', text='Média de Avaliaçoes', 
                 title='Média de Avaliaçoes: restaurantes com e sem pedidos online', 
                 color='Tipo', color_discrete_map={ 'Online': 'green', 'Offline': 'red'}
)
    fig.update_layout(title_x=0.5)
    fig.update_traces(textposition='outside')

    return fig

grafico_online_offline(df)
```

```python
def grafico_japao_bbq(df):
    japonesa = df.loc[(df['cuisines'] == 'Japanese') & (df['country_name'] == 'United States of America') , 'average_cost_for_two' ].mean().round(1)
    bbq = df.loc[(df['cuisines'] == 'BBQ') & (df['country_name'] == 'United States of America') , 'average_cost_for_two'].mean().round(1)

    grafico = pd.DataFrame( {
        'Tipo': ['Japonesa', 'BBQ'],
        'Valor Medio Prato para Duas Pessoas': [japonesa, bbq]
    })

    fig = px.bar(grafico, x='Tipo', y='Valor Medio Prato para Duas Pessoas', 
                 title='Valor Médio de Prato para Duas Pessoas: Japonesa x BBQ', 
                 color='Tipo', color_discrete_map={'Japonesa': 'red', 'BBQ': 'orange'},
                 text='Valor Medio Prato para Duas Pessoas')
    
    fig.update_layout(title_x=0.5)
    fig.update_layout(yaxis_title='Valor Médio')
    fig.update_traces(textposition='outside')
    
    return fig

grafico_japao_bbq(df)
```
