# Perguntas de Negócio

### 1. Qual o nome da cidade que possui mais restaurantes registrados?

```python
cidade_mais_restaurantes = (df.groupby('city')
                            .agg(
                                quantidade_restaurantes=('restaurant_id', 'nunique'),
                                menor_id=('restaurant_id', 'min')
                            ).sort_values(by=['quantidade_restaurantes', 'menor_id'], ascending=[False, True]).reset_index().iloc[0])
nome = cidade_mais_restaurantes['city']
qtd_restaurantes = cidade_mais_restaurantes['quantidade_restaurantes']

print(f'A cidade com mais restaurantes é {nome}, com {qtd_restaurantes} registrados')
```

### 2. Qual o nome da cidade que possui mais restaurantes com nota média acima de 4?

```python
restaurante_nota_alta = df.loc[df['aggregate_rating'] > 4 , ['city' , 'restaurant_id']].groupby('city')['restaurant_id'].count().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').iloc[0]
nome = restaurante_nota_alta['city']
qtd_cidades = restaurante_nota_alta['Quantidade de Restaurantes']

print(f'A cidade com mais restaurantes com nota média acima de 4 é {nome}, com {qtd_cidades} restaurantes')
```

### 3. Qual o nome da cidade que possui mais restaurantes com nota média abaixo de 2.5?

```python
restaurante_nota_baixa = df.loc[df['aggregate_rating'] < 2.5 , ['city' , 'restaurant_id']].groupby('city')['restaurant_id'].count().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').iloc[0]
nome = restaurante_nota_baixa['city']
qtd_restaurantes = restaurante_nota_alta['Quantidade de Restaurantes']

print(f'A cidade com mais restaurantes com nota média abaixo de 2.5 é {nome}, com {qtd_restaurantes} restaurantes')
```

### 4. Qual o nome da cidade que possui o maior valor médio de um prato para dois?

```python
maior_valor_prato_dois = df.loc[: , ['city', 'average_cost_for_two']].groupby('city')['average_cost_for_two'].mean().sort_values(ascending=False).reset_index(name='Preco Médio para Dois').iloc[0]
nome = maior_valor_prato_dois['city']
preco_medio = maior_valor_prato_dois['Preco Médio para Dois'].round(2)

print(f'A cidade com o maior valor médio de um prato para dois é {nome}, com um preço médio de {preco_medio}')
```

### 5. Qual o nome da cidade que possui a maior quantidade de tipos de culinária distintas?

```python
cidade_culinaria_distinta = df.loc[: , ['city', 'cuisines']].groupby('city')['cuisines'].nunique().sort_values(ascending=False).reset_index(name='Tipos de Culinarias Distintas').iloc[0]
nome = cidade_culinaria_distinta['city']
qtd_culinarias = cidade_culinaria_distinta['Tipos de Culinarias Distintas']

print(f'A cidade com a maior quantidade de tipos de culinária distintas é {nome}, com {qtd_culinarias} tipos')
```

### 6. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem reservas?

```python
df_reserva = df.loc[ df['has_table_booking'] == 1 , ['city', 'restaurant_id']].groupby('city')['restaurant_id'].nunique().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').iloc[0]
nome = df_reserva['city']
qtd_restaurantes = df_reserva['Quantidade de Restaurantes']

print(f'A cidade com a maior quantidade de restaurantes que fazem reserva é {nome}, com {qtd_restaurantes} restaurantes')
```

### 7. Qual o nome da cidade que possui a maior quantidade de restaurantes que fazem entregas?

```python
df_entrega = (df.loc[ df['is_delivering_now'] == 1  , : ].groupby('city')
              .agg(
                  quantidade_de_restaurantes=('restaurant_id', 'nunique'),
                  menor_ID=('restaurant_id', 'min')
              )  .sort_values(by=['quantidade_de_restaurantes', 'menor_ID'], ascending=[False, True]).reset_index().iloc[0])
nome = df_entrega['city']
qtd_restaurantes = df_entrega['quantidade_de_restaurantes']

print(f'A cidade com a maior quantidade de restaurantes que fazem entrega é {nome}, com {qtd_restaurantes} restaurantes')
```

# Funções de Gráficos

```python
# Top 10 cidades com mais restaurantes

def top_cidades_restaurantes(df):
    cidade_mais_restaurantes = (df.groupby('city')
                                .agg(
                                    quantidade_restaurantes=('restaurant_id', 'nunique'),
                                    menor_id=('restaurant_id', 'min')
                                ).sort_values(by=['quantidade_restaurantes', 'menor_id'], ascending=[False, True]).reset_index().head(10))
    fig = px.bar(cidade_mais_restaurantes, x='city', y='quantidade_restaurantes', text='quantidade_restaurantes', color='city', labels={'city': 'Cidades'})
    
    fig.update_layout(xaxis_title = 'Cidades', showlegend=False)
    fig.update_traces(textposition='outside')
    
    return fig

top_cidades_restaurantes(df)
```

```python
# Top 7 cidades com restaurantes com media acima de 4

def top_restaurantes(df):
    restaurante_nota_alta = df.loc[df['aggregate_rating'] > 4 , ['city' , 'restaurant_id']].groupby('city')['restaurant_id'].count().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').head(7)
    fig = px.bar(restaurante_nota_alta, x='city', y='Quantidade de Restaurantes', text='Quantidade de Restaurantes', color='city', labels={'city': 'Cidades'})
    
    fig.update_layout(xaxis_title = 'Cidades', showlegend=False)
    fig.update_traces(textposition='outside')
    
    return fig

top_restaurantes(df)
```

```python
# Top 7 cidades com restaurantes com media abaixo de 2.5

def piores_restaurantes(df):
    restaurante_nota_baixa = df.loc[df['aggregate_rating'] < 2.5 , ['city' , 'restaurant_id']].groupby('city')['restaurant_id'].count().sort_values(ascending=False).reset_index(name='Quantidade de Restaurantes').head(7)
    fig = px.bar(restaurante_nota_baixa, x='city', y='Quantidade de Restaurantes', text='Quantidade de Restaurantes', color='city', labels={'city': 'Cidades'})
    
    fig.update_layout(xaxis_title = 'Cidades', showlegend=False)
    fig.update_traces(textposition='outside')
    
    return fig

piores_restaurantes(df)
```
