# Limpeza dos Dados

```python
# Criando Funções:

COUNTRIES = {
1: "India",
14: "Australia",
30: "Brazil",
37: "Canada",
94: "Indonesia",
148: "New Zeland",
162: "Philippines",
166: "Qatar",
184: "Singapure",
189: "South Africa",
191: "Sri Lanka",
208: "Turkey",
214: "United Arab Emirates",
215: "England",
216: "United States of America",
}

def Country_Name(country_id):
    return COUNTRIES[country_id]


# -----------------------------------------------


def create_price_tye(price_range):
    if price_range == 1:
        return "cheap"
    elif price_range == 2:
        return "normal"
    elif price_range == 3:
        return "expensive"
    else:
        return "gourmet"


# -----------------------------------------------

COLORS = {
"3F7E00": "darkgreen",
"5BA829": "green",
"9ACD32": "lightgreen",
"CDD614": "orange",
"FFBA00": "red",
"CBCBC8": "darkred",
"FF7800": "darkred",
}
def color_name(color_code):
    return COLORS[color_code]

# -----------------------------------------------

def rename_columns(dataframe):
    df = dataframe.copy()
    title = lambda x: inflection.titleize(x)
    snakecase = lambda x: inflection.underscore(x)
    spaces = lambda x: x.replace(" ", "")
    cols_old = list(df1.columns)
    cols_old = list(map(title, cols_old))
    cols_old = list(map(spaces, cols_old))
    cols_new = list(map(snakecase, cols_old))
    df1.columns = cols_new
    return df
```

```python
# Define categorias de preço de acordo com o range
df['price_category'] = df['Price range'].apply(create_price_tye)

# Define o padrão de cores das avaliações
df['rating_color'] = df['Rating color'].apply(color_name)

# --------------------------------------------------------

df['country_name'] = df['Country Code'].map(Country_Name)
# Definisdo os restaurantes po apenas um tipo de culinaria
df["Cuisines"] = df.loc[:, "Cuisines"].astype(str).apply(lambda x: x.split(",")[0])

# Deletando coluna desnecessária, todos os valores são 0
df = df.drop(['Switch to order menu'], axis=1)

# Deixando todas as colunas em formato snake_case

df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')

# --------------------------------------------------------

# Verificar quanto em % representa os dados nulos

df.isna().mean() * 100

# Dropando todos os dados nulos

df = df.dropna()
```

# Perguntas de Negócio

### 1. Dos restaurantes que possuem o tipo de culinária italiana, qual o nome do restaurante com a maior média de avaliação?


```python
df_italiana = (df.loc[ df['cuisines'] == 'Italian' , ['restaurant_name' , 'aggregate_rating', 'restaurant_id']].groupby('restaurant_name')
               .agg(
                   maior_media=('aggregate_rating', 'max'),
                   menor_id=('restaurant_id', 'min')
               ) .sort_values(by= ['maior_media', 'menor_id'],ascending=[False, True]).reset_index().iloc[0])
nome = df_italiana['restaurant_name']
maior_media = df_italiana['maior_media']

print(f'O restaurante com a culinária do tipo italiana e com a maior média é {nome}, com {maior_media}')
```

# Funções e Gráficos

```python
# Top 10 melhores restaurantes

def top_restaurantes(df):
    cols = ['restaurant_id','restaurant_name','country_name','city','cuisines','average_cost_for_two','aggregate_rating','votes']

    top = (df.loc[:, cols].groupby('restaurant_name')
           .agg({'restaurant_id':'min','country_name':'first','city':'first','cuisines':'first',
                 'average_cost_for_two':'first','aggregate_rating':'mean','votes':'sum'})
           .reset_index()
           .sort_values(by=['aggregate_rating','votes','restaurant_id'],
                        ascending=[False, False, True])
           .head(10))

    return top

top_restaurantes(df)
```

```python
# Top 10 melhores culinarias
def top_culinarias(df):
    culinaria_nota_media = (df.loc[ : , ['cuisines', 'aggregate_rating', 'restaurant_id', 'country_name']].groupby('cuisines')
                            .agg(
                                maior_nota = ('aggregate_rating', 'max'),
                                menor_id = ('restaurant_id', 'min'),
                                country_name=('country_name', 'first')
                            ).sort_values(by=['maior_nota', 'menor_id'] , ascending=[False, True]).reset_index().head(10))
    
    fig = px.bar(culinaria_nota_media, x='cuisines', y='maior_nota', text='maior_nota', labels={'country_name': 'Países'} )
    fig.update_layout(
                      xaxis_title='Culinária',
                      yaxis_title='Média Nota'
                      )
    
    fig.update_traces(textposition='outside')

    return fig

top_culinarias(df)
```
